```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
butterfly = cv.imread("/content/drive/MyDrive/Colab Notebooks/EN 2550/butterfly.jpg",cv.IMREAD_REDUCED_COLOR_4)
assert butterfly is not None

k_size = 9
sigma = 4
box_kernal = 1./81*np.ones((k_size,k_size))
im_avg = cv.filter2D(butterfly,-1,box_kernal)
im_Gaussian = cv.GaussianBlur(butterfly,(k_size,k_size),sigma)

print("Original Image")
cv2_imshow(butterfly)
print("Box Filtered")
cv2_imshow(im_avg)
print("Gaussian Blured Image")
cv2_imshow(im_Gaussian)
```

```python
#Q 3
lens = cv.imread("/content/drive/MyDrive/Colab Notebooks/EN 2550/contact_lens.tif",cv.IMREAD_GRAYSCALE).astype(np.float32)
assert butterfly is not None

kernal_vertical = np.array([[-1,-2,-1], [0,0,0], [1,2,1]], dtype=float)
kernal_horizontal = kernal_vertical.transpose()

f_y = cv.filter2D(lens,-1,kernal_vertical)
f_x = cv.filter2D(lens,-1,kernal_horizontal)
grad_im = np.sqrt(f_y**2 + f_x**2)

print("Original Image")
cv2_imshow(lens)
print("Sobel Vertical")
cv2_imshow(f_y)
print("Sobel Horizontal")
cv2_imshow(f_x)
print("Grad Image")
cv2_imshow(grad_im)
```

```python
# using matplotlib
fig,axes =plt.subplots(1 ,4 ,figsize = (20,18) )

axes[0].imshow(lens , cmap=  'gray', vmin = 0, vmax = 255)
axes[0].set_title('Original')

axes[1].imshow(f_y  cmap=  'gray', vmin = 0, vmax = 255)
axes[1].set_title('Sobel Vertical')

axes[2].imshow(f_x , cmap=  'gray' , vmin = -1020 , vmax = 1020)
axes[2].set_title('Sobel Horizontal')

axes[3].imshow(grad_im , cmap=  'gray' )
axes[3].set_title('Gradient Magnitude')
```

```python
# Q 4

tom = cv.imread("/content/drive/MyDrive/Colab Notebooks/EN 2550/tom.jpg",cv.IMREAD_GRAYSCALE).astype(np.float32)
assert butterfly is not None

sigma = 2
gaussian_1D = cv.getGaussianKernel(5,sigma)
f_LP = cv.sepFilter2D(tom,-1,gaussian_1D,gaussian_1D)
f_HP = tom - f_LP
f_sharp = cv.addWeighted(tom,1.0,f_HP,2.0,0)

print("Original Image")
cv2_imshow(tom)
print("Low Passed Image")
cv2_imshow(f_LP)
print("High Passed Image")
cv2_imshow(f_HP)
print("Sharpned Image")
cv2_imshow(f_sharp)
```

```python
# using matplotlib
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1,4, figsize=(20,10))
axes[0].imshow(tom,cmap = 'gray')
axes[0].set_title("Original")

axes[1].imshow(f_LP,cmap = 'gray')
axes[1].set_title("low pass")

axes[2].imshow(f_HP,cmap = 'gray')
axes[2].set_title("high pass")

axes[3].imshow(f_sharp,cmap = 'gray')
axes[3].set_title("sharp")
for i in range(4):
  axes[i].axis("off")

plt.show()
```

```python

#Q 2
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
#from matplotlib.ticker import LinearLocator, FormatStrFormatter

fig,ax = plt.subplots()
ax = fig.add_subplot(111, projection="3d")

step = 0.1
X = np.arange(-5, 5 + step, step)
Y = np.arange(-5, 5 + step, step)
XX, YY = np.meshgrid(X,Y)
sigma = 1
g = np.exp(-(XX**2 + YY**2)/(2*sigma**2))

surf = ax.plot_surface(XX,YY , g ,cmap=cm.jet)

cset = ax.contourf(XX,YY,g, zdir="z",offset=np.min(g)-1.5,cmap=cm.jet)
ax.set_zlim(np.min(g)-2, np.max(g))
plt.axis("off")

plt.show()
```
